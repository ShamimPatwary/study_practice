### Advance Operation With Business Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure : [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],
    [2, 120000, 140000, 160000, 190000],
    [3, 200000, 230000, 260000, 300000],
    [4, 180000, 210000, 240000, 270000],
    [5, 160000, 185000, 205000, 230000]
])

print("Foodpanda sales analysis: ")
print("\nSales data shapes : ", sales_data.shape)
print("\nSample data for 1st 3 restaurant : \n", sales_data[0:3])
```

```python
# total sales per year
print(np.sum(sales_data, axis=0))
per_year = np.sum(sales_data[:, 1:], axis=0)
print(per_year)


# minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)

# maximum sales per restaurant
max_sales = np.max(sales_data[:, 1:], axis=0)
print(max_sales)

# average sals per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales)
```

```python
cumalitive_sum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumalitive_sum)

plt.figure(figsize=(14,6))
plt.plot(np.mean(cumalitive_sum, axis=0))
plt.title("Average Cumalative Sum Accross All Restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector addition : ", vector1 + vector2)

print("Vector multiplications :", vector1 * vector2)

print("Dot product : ", np.dot(vector1 , vector2))

angle = np.angle(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```
