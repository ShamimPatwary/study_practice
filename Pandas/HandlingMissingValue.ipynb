NAN (Not a Number)
None (for object data types)

isnull() -- method 

True -> Nan is missing
and False -> Value is present

```python
import pandas as pd
```

```python
data = {
    "Name":['shamim', None, 'sharif', 'patwary', 'tafin', 'sadid'],
    "Age":[22, 26, 50,None, 13, 11],
    "Salary" :[5000, 10000, None, 20000, 3000, 2000],
    "Performance_Score":[90, 95, 88, 80, None, 75]
}

df = pd.DataFrame(data)

print(df.isnull())
print(df.isnull().sum())
```

### Handling

```python
### Removing Missing Value
df.dropna(axis=0, inplace=False)  ## axis = 0 -> Row / axis = 1 -> Column
print(df)
```

```python
### Without removing missing value
df["Name"].fillna("Unknown", inplace=False)
df[["Age","Salary","Performance_Score"]].fillna(0, inplace=False)

print(df)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=False)
print(df)
```
