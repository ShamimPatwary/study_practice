### Dive into ###

```python
data = {
    "Name":['shamim', 'sharif', 'patwary', 'tafin', 'sadid'],
    "Age":[22, 26, 50, 13, 11],
    "Salary" :[5000, 10000, 20000, 3000, 2000],
    "Performance_Score":[90, 95, 88, 80, 75]
}

df3 = pd.DataFrame(data)
print("Simple data frame\n")
print(df3)

print("\nSelecting single column\n") ##Single column return series
print(df3['Name'])

print("\nSelecting multiple column")
subset = df3[['Name', 'Salary']]
print(subset)

##Filtering column
print("Salary > 5000")
high_salary = df3[df3['Salary'] > 5000]
print(high_salary)

##Filtering age and salary
print("Age > 20 and salary > 5000")
filtered = df3[(df3['Age'] > 20) & (df3['Salary'] > 5000)]
print(filtered)
```
