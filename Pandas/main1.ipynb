### Adding Colums

```python
import pandas as pd
```

```python
data = {
    "Name":['shamim', 'sharif', 'patwary', 'tafin', 'sadid'],
    "Age":['22', '26', '50', '13', '11'],
    "Salary" :[5000, 10000, 20000, 3000, 2000],
    "Performance_Score":[90, 95, 88, 80, 75]
}

df = pd.DataFrame(data)
```

```python
###Adding column method 1
df['Bonus'] = df['Salary'] * 0.1
print(df)
```

```python
###Adding column with insert method
df.insert(0, "Id", [1,2,3,4,5])
print(df)
```

### Updating Value

```python
df.loc[0, "Salary"] = 6000
print(df)
```

```python
df['Salary'] = df['Salary'] * 1.05
print(df)
```

### Removing Column and Rows

```python
df.drop(columns=['Age', 'Salary'], inplace=True)
print(df)
```
